How many power plants are eligible for California's
Renewable Portfolio Standards program?

```python
cpp['RPS_Eligible'].value_counts()
```

It looks like 1353 power plants are eligible while 498 are not. Let's save the RPS Eligibility counts as a variable.

```python
RPS_Count = cpp['RPS_Eligible'].value_counts()
RPS_Count
```

Let's make this a dataframe

```python
RPS_Count = RPS_Count.reset_index()
RPS_Count
```

Let's go back to the original dataframe and use the .loc feature to only see which plants are RPS eligible (assigned a value of 1.0 under the column RPS_Eligible)

```python
cpp.loc[cpp['RPS_Eligible'] == 1.0]
```

Let's save rps_eligible as a variable

```python
rps_eligible = cpp.loc[cpp['RPS_Eligible'] == 1.0]
rps_eligible
```

How many RPS eligible plants are in the county of Los Angeles?

```python
LA_rps_eligible = rps_eligible.loc[rps_eligible['County'] == 'Los Angeles']
LA_rps_eligible
```
